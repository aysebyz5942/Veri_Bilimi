# Aykırı veri Bulma

```python
q75, q25 = np.percentile(df.Aspect, [75 ,25])
IQR = q75 - q25
threshold_values=pd.DataFrame(columns={'threshold','outliers_number'})
for threshold in np.arange(1,4,0.5):
    min_value = q25 - (IQR*threshold)
    max_value = q75 + (IQR*threshold)
    outliers_number = len((np.where((df.Aspect > max_value)| (df.Aspect < min_value))[0]))
    threshold_values = threshold_values.append({'threshold': threshold,
                                                'outliers_number': outliers_number},ignore_index=True)
threshold_values
```

# data smoothing with data minning

```python
import pandas as pd
import numpy as np
import math
from sklearn.datasets import load_iris
from sklearn import datasets, linear_model, metrics

a = df.values
b = np.zeros(126426)

# take 1st column among 4 column of data set
for i in range (126426):
	b[i]=a[i,1]

b=np.sort(b) #sort the array

# create bins
bin1=np.zeros((126426,56))
bin2=np.zeros((126426,56))
bin3=np.zeros((126426,56))

# Bin mean
for i in range (0,126426,56):
	k=int(i/56)
	mean=(b[i] + b[i+1] + b[i+2] + b[i+3] + b[i+4])/56
	for j in range(56):
		bin1[k,j]=mean
print("Bin Mean: \n",bin1)
	

# Bin median
for i in range (0,126426,56):
	k=int(i/56)
	for j in range (56):
		bin3[k,j]=b[i+2]
print("Bin Median: \n",bin3)
```
