```python
kmeans=KMeans(2) #k=2 olsun dedik sadece 2 kümeye ayıracak
kmeans.fit(x) #model oluşturuldu
```

```python
clusters=x.copy()
clusters['cluster_pred']=kmeans.fit_predict(x) # x verisine gore model oluşturduk
```

```python
from sklearn import preprocessing
x_scaled=preprocessing.scale(x)
x_scaled
#dirsek yöntemi ile daha güzel bir kümeleme yapmak istedim bu yüzden tekrar denedim
```

# Dirsek Yöntemi

```python
wcss=[] #liste oluşturduk

for i in range(1,10): #10 örnek olduğu için 
    kmeans=KMeans(i)
    kmeans.fit(x_scaled) #standartlaştığımız datamııza göre fit et, yani modelimizi oluştur
    wcss.append(kmeans.inertia_) #her bir k değeri için wcss değerini bul
```

```python
#dirsek grafiğini oluşturarak dirseğe benzeyen yerdeki değerini kullanacağım kümeleme sayısı için, bu da 2 noktası oluyor.
plt.scatter(range(1,10),wcss)
plt.xlabel('K_değer')
plt.ylabel('WCSS')
plt.show
```

```python
#DBSCAN ile 2 kümeden oluşan bir eğitim yaptım
db = DBSCAN(eps=2, min_samples=10).fit(df2)
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
labels = db.labels_
```

```python
# clusters label'ların sayısı,gürültüyü işleme almadık
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

print("Estimated number of clusters: %d" % n_clusters_) #kümelenmiş sayısı
print("Estimated number of noise points: %d" % n_noise_) #gürültü sayısı
```

```python
# Grafiğin sonucu

unique_labels = set(labels)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Siyah noktalar gürültüyü gösterir
        col = [0, 0, 0, 1]

    class_member_mask = labels == k

    xy = df2[class_member_mask & core_samples_mask]
    plt.plot(
        xy[:, 0],
        xy[:, 1],
        "o",
        markerfacecolor=tuple(col),
        markeredgecolor="k",
        markersize=14,
    )

    xy = df2[class_member_mask & ~core_samples_mask]
    plt.plot(
        xy[:, 0],
        xy[:, 1],
        "o",
        markerfacecolor=tuple(col),
        markeredgecolor="k",
        markersize=6,
    )

plt.title("Estimated number of clusters: %d" % n_clusters_)
plt.show()
```
